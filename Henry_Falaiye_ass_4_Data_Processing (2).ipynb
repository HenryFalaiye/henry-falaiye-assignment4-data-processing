```python
df_null = pd.DataFrame({
    "Name": ["John", None, "Peter"],
    "Age": [25, np.nan, 28],
    "Salary": [50000, None, 60000]
})

print("Original DataFrame:")
print(df_null)

# Fill nulls
df_filled = df_null.fillna({
    "Name": "Unknown",
    "Age": df_null["Age"].mean(),
    "Salary": df_null["Salary"].mean()
})

print("\nFilled DataFrame:")
print(df_filled)
```

```python
df1 = pd.DataFrame({"A": [1, 2], "B": [3, 4]})
df2 = pd.DataFrame({"A": [5, 6], "B": [7, 8]})

# Concatenate
concat_df = pd.concat([df1, df2], axis=0)
print(concat_df)

df3 = pd.DataFrame({"A": [1, 2, 5, 6], "C": ["X", "Y", "Z", "W"]})

# Merge
merged_df = pd.merge(concat_df, df3, on="A")
print(merged_df)
```

```python
# Iterator
class EvenNumbers:
    def __init__(self):
        self.num = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            result = self.num
            self.num += 2
            self.count += 1
            return result
        else:
            raise StopIteration

for i in EvenNumbers():
    print(i)

# Generator
def even_generator():
    for i in range(0, 10, 2):
        yield i

for i in even_generator():
    print(i)
```

```python
from functools import reduce

nums = [1, 2, 3, 4]

# Map
print(list(map(lambda x: x**2, nums)))

# Reduce
print(reduce(lambda x, y: x*y, nums))

# Filter
print(list(filter(lambda x: x % 2 == 0, nums)))
```

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

r1 = Rectangle(5, 3)
r2 = Rectangle(10, 4)

print(r1.area(), r1.perimeter())
print(r2.area(), r2.perimeter())
```
