```python
import pandas as pd
import numpy as np

data = {
    "Name": ["John", "Mary", "Peter"],
    "Age": [25, 30, 28],
    "City": ["New York", "London", "Toronto"]
}

df = pd.DataFrame(data)
print(df)
print(df.dtypes)
```

```python
df2 = df.drop(columns=["City"])
print(df2)
```

```python
df_null = pd.DataFrame({
    "Name": ["John", None, "Peter"],
    "Age": [25, np.nan, 28],
    "Salary": [50000, None, 60000]
})

print("Original DataFrame:")
print(df_null)

# Fill nulls
df_filled = df_null.fillna({
    "Name": "Unknown",
    "Age": df_null["Age"].mean(),
    "Salary": df_null["Salary"].mean()
})

print("\nFilled DataFrame:")
print(df_filled)
```

```python
df_group = pd.DataFrame({
    "Category": ["A", "A", "B", "B", "C"],
    "Value": [10, 20, 15, 25, 30]
})

grouped = df_group.groupby("Category")["Value"].describe()
print(grouped)
```

```python
df1 = pd.DataFrame({"A": [1, 2], "B": [3, 4]})
df2 = pd.DataFrame({"A": [5, 6], "B": [7, 8]})

# Concatenate
concat_df = pd.concat([df1, df2], axis=0)
print(concat_df)

df3 = pd.DataFrame({"A": [1, 2, 5, 6], "C": ["X", "Y", "Z", "W"]})

# Merge
merged_df = pd.merge(concat_df, df3, on="A")
print(merged_df)
```
